```python
import gradio as gr
from transformers import pipeline

# Load the BART model for summarization
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text(text):
    # Check if the input text is empty or only contains spaces
    if not text.strip():
        return "❌ Please enter article text."

    # Split the text into chunks of 1024 characters to fit within the model's input limit
    chunks = [text[i:i+1024] for i in range(0, len(text), 1024)]
    summary = ""

    # Summarize each chunk and combine the summaries
    for chunk in chunks:
        result = summarizer(chunk, max_length=130, min_length=30, do_sample=False)
        summary += result[0]['summary_text'] + " "

    # Return the concatenated summary
    return summary.strip()

# Gradio UI setup for text summarization
with gr.Blocks() as summarizer_text_ui:
    # Markdown header
    gr.Markdown("## 📰 AI News Summarizer (Paste Article Text)")

    # Textbox for article input
    article_input = gr.Textbox(lines=10, label="Paste News Article Text")

    # Button to trigger summarization
    summarize_btn = gr.Button("Summarize")

    # Textbox to display the summarized output
    summary_output = gr.Textbox(label="Summary")

    # Connect button to function (summarize_text) to trigger on click
    summarize_btn.click(summarize_text, inputs=article_input, outputs=summary_output)

# Launch the Gradio interface and share the link
summarizer_text_ui.launch(share=True)
```
